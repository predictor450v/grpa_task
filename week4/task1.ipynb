```python
singleton_list = [1] # list: A list with only one element
singleton_set = {1} # set: A set with only one element
singleton_tuple = (1,) # tuple: A tuple with only one element
```

```python
if len(singleton_list) == 1:
    print("List has exactly 1 element")

if len(singleton_set) == 1:
    print("Set has exactly 1 element")

if len(singleton_tuple) == 1:
    print("Tuple has exactly 1 element")
```
